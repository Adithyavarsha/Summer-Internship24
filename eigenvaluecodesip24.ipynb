```python
import sympy as sp

# Define the variables
a, f, k, p, g, l, n, h, j, o, b, e, i, c, d, m, z = sp.symbols('a f k p g l n h j o b e i c d m z')

# Define the polynomial expression
polynomial = ((a - z)*(f - z)*(k - z)*(p - z) + (a - z)*g*l*n + (a - z)*h*j*o - (a - z)*h*(k - z)*n
              - (a - z)*g*i*(p - z) - (a - z)*(f - z)*l*o - b*e*(k - z)*(p - z) - c*e*l*n
              - d*e*j*o + d*e*(k - z)*n + c*e*j*(p - z) + b*e*l*o + b*g*i*(p - z)
              + c*h*i*n + d*(f - z)*i*o - d*g*i*n - c*(f - z)*i*(p - z) - b*h*i*o
              - b*g*l*m - c*h*j*m - d*(f - z)*(k - z)*m + d*g*j*m + c*(f - z)*j*m
              + b*h*(k - z)*m)

# Expand the polynomial
expanded_poly = sp.expand(polynomial)

# Collect coefficients for each power of z
coefficients = sp.Poly(expanded_poly, z).all_coeffs()

print(coefficients)
```

```python
import sympy as sp

# Define the variable and coefficients
z = sp.symbols('z')
a, b, c, d, e = sp.symbols('a b c d e')

# Define the quartic polynomial
quartic_polynomial = a*z**4 + b*z**3 + c*z**2 + d*z + e

# Solve the polynomial equation P(z) = 0
solutions = sp.solve(quartic_polynomial, z)

print(solutions)
```
