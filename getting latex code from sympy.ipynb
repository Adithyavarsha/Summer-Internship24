```python
import sympy as sp

# Define the variable and coefficients
z = sp.symbols('z')
a, b, c, d, e = sp.symbols('a b c d e')

# Define the quartic polynomial
quartic_polynomial = z**4 + a*z**3 + b*z**2 + c*z + d

# Solve the polynomial equation P(z) = 0
solutions = sp.solve(quartic_polynomial, z)

latex_code = sp.latex(solutions)
print(latex_code)
```
