```python
from sympy import symbols, solve, latex

# Define the symbols
A,B,C,D,z = symbols('A B C D z')

# Set up the equation
equation= z**4+A*z**3+B*z**2+C*z+D
# Solve the equation for z
solutions = solve(equation, z)
print(solutions)
latex_code=latex(solutions)
print("LaTeX code")
print(latex_code)
```
